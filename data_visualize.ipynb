```python
import tensorflow as tf
from tensorflow.keras import models,layers
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_size= 200
Batch_size=32
epochs=5
data_folder ='dataset'
channels=3
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    data_folder,
    shuffle=True,
    image_size=(image_size,image_size),
    batch_size=Batch_size
)
```

```python
class_names=dataset.class_names
class_names
```

```python
len(dataset)
```

## Understanding the data
    1 batch has 32 images where the channel is 3 which is rgb and size is 200*200
    and label batch is the labels for the images.

```python
for image_batch, label_batch in dataset.take(1):
    print(image_batch.shape)
    print(label_batch.numpy())
```

```python
for image_batch in dataset.take(1):
    ## First batches first image
    print(image_batch[0].numpy())
```

### Visualizing the data
    Shuffle is turned on so when you run it again and again it will give new image

```python
for image_batch, label_batch in dataset.take(1):
    plt.imshow(image_batch[0].numpy().astype('uint8'))
    plt.axis('off')
```

#### Stacking images 

```python
plt.figure(figsize=(10,8))
for image_batch, label_batch in dataset.take(1):
    for i in range(8):
        ax= plt.subplot(4,4,i+1)
        plt.imshow(image_batch[i].numpy().astype('uint8'))
        plt.title(class_names[label_batch[i]])
        plt.axis('off')
```
